Objetivo 3. Analizar la proporción de establecimientos educativos con matrícula contratada (paga) frente a los que no la tienen, por zona y departamento.

```python
# Asegurar consistencia
df_cleaned['matricula_Contratada'] = df_cleaned['matricula_Contratada'].str.upper().fillna('NO')

# Filtrar los que sí tienen matrícula contratada
establecimientos_contratada = df_cleaned[df_cleaned['matricula_Contratada'] == 'SI']

# Contar cuántos hay por departamento
contratada_por_dep = establecimientos_contratada.groupby('nombredepartamento').size().reset_index(name='con_matricula_contratada')
```

```python
# Unir totales y contratadas
cobertura_matricula = pd.merge(establecimientos_totales, contratada_por_dep, on='nombredepartamento', how='left')

# Llenar con 0 donde no hay matrícula contratada
cobertura_matricula['con_matricula_contratada'] = cobertura_matricula['con_matricula_contratada'].fillna(0).astype(int)

# Calcular porcentaje
cobertura_matricula['porcentaje_contratada'] = (
    cobertura_matricula['con_matricula_contratada'] / cobertura_matricula['total_establecimientos'] * 100
).round(2)

print(cobertura_matricula.sort_values(by='porcentaje_contratada', ascending=False))
```

Visualizaciones del objetivo 3: Analizar la proporción de establecimientos educativos que cuentan con matrícula contratada frente a los que no, según departamento y zona.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Normaliza los textos
df_cleaned['zona'] = df_cleaned['zona'].str.strip()
df_cleaned['matricula_Contratada'] = df_cleaned['matricula_Contratada'].str.strip().str.lower()

# Agrupa por departamento, zona y matrícula contratada
conteo = df_cleaned.groupby(['nombredepartamento', 'zona', 'matricula_Contratada']).size().reset_index(name='conteo')

# Crea columna para usar como clave de color
conteo['grupo'] = conteo['zona'] + ' - ' + conteo['matricula_Contratada']

# Tabla pivote para gráfico apilado
pivot = conteo.pivot_table(index='nombredepartamento', columns='grupo', values='conteo', fill_value=0)

# Ordenar departamentos por total de matrícula contratada
columnas_contratada = [col for col in pivot.columns if col.endswith('sí')]
pivot['total_contratada'] = pivot[columnas_contratada].sum(axis=1)
pivot = pivot.sort_values(by='total_contratada', ascending=False).drop(columns='total_contratada')

# Colores vibrantes conocidos
colores_personalizados = {
    'Urbana - sí': '#e6194B',     # rojo fuerte
    'Urbana - no': '#f58231',     # naranja vibrante
    'Rural - sí': '#3cb44b',      # verde fuerte
    'Rural - no': '#aaffc3',      # verde claro vibrante
}

# Detectar claves faltantes
faltantes = [col for col in pivot.columns if col not in colores_personalizados]

# Colores vibrantes para asignar a claves faltantes
colores_vibrantes = [
    '#911eb4',  # púrpura
    '#46f0f0',  # cian claro
    '#f032e6',  # rosa brillante
    '#fabebe',  # rosa claro
    '#008080',  # teal
    '#e6beff',  # lavanda claro
    '#9A6324',  # marrón (repetido, pero puede usar otro)
    '#ffd8b1',  # melocotón claro
]

# Asignar colores a faltantes en orden, repitiendo si es necesario
for i, col in enumerate(faltantes):
    colores_personalizados[col] = colores_vibrantes[i % len(colores_vibrantes)]

# Gráfico
plt.figure(figsize=(14, 12))
pivot.plot(
    kind='barh',
    stacked=True,
    figsize=(14, 12),
    color=[colores_personalizados.get(c, '#000000') for c in pivot.columns]  # negro si algo falla
)

plt.title('Número de Establecimientos con y sin Matrícula Contratada por Zona y Departamento')
plt.xlabel('Número de Establecimientos')
plt.ylabel('Departamento')
plt.legend(title='Zona - Matrícula Contratada', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Importar librerías
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Leer archivo GeoJSON actualizado
url = 'https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/3aadedf47badbdac823b00dbe259f6bc6d9e1899/colombia.geo.json'
gdf_departamentos = gpd.read_file(url)

# Asegurar formato de nombres en tu dataframe
df['nombredepartamento'] = df['nombredepartamento'].str.lower().str.strip()
# Use the correct column name 'NOMBRE_DPT' instead of 'Departamento'
gdf_departamentos['NOMBRE_DPT'] = gdf_departamentos['NOMBRE_DPT'].str.lower().str.strip()

# Agrupar datos por departamento
establecimientos_por_dep = df.groupby('nombredepartamento').size().reset_index(name='conteo_establecimientos')

# Unir datos con el GeoDataFrame
gdf_merged = gdf_departamentos.merge(establecimientos_por_dep, left_on='NOMBRE_DPT', right_on='nombredepartamento', how='left')

# Llenar vacíos con 0
gdf_merged['conteo_establecimientos'] = gdf_merged['conteo_establecimientos'].fillna(0)

# Crear el mapa
plt.figure(figsize=(16, 12))
ax = gdf_merged.plot(column='conteo_establecimientos',
                     cmap='YlGnBu',
                     linewidth=0.8,
                     edgecolor='black',
                     legend=True,
                     figsize=(16, 12))

# Añadir nombres de departamentos
for idx, row in gdf_merged.iterrows():
    # Use the correct column name 'NOMBRE_DPT' for annotation
    plt.annotate(text=row['NOMBRE_DPT'].title(), xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                 horizontalalignment='center', fontsize=8, color='black')

# Estética del gráfico
plt.title('Cobertura de Establecimientos Educativos por Departamento en Colombia', fontsize=16)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import folium
import matplotlib.pyplot as plt

# Se Cargan lo datos
file_path = "/cleaned_ESTABLECIMIENTOS_EDUCATIVOS-COLOMBIA_20250605 (1) (6).csv" # Replace with the correct file path
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found. Please provide the correct path to your CSV file.")

#Se  Normalizan columnas
df['nombredepartamento'] = df['nombredepartamento'].str.lower().str.strip()
df['nombremunicipio'] = df['nombremunicipio'].str.lower().str.strip()
df['matricula_Contratada'] = df['matricula_Contratada'].str.lower().str.strip()

# Se Crea variable binaria
df['con_matricula_contratada'] = df['matricula_Contratada'].apply(lambda x: 1 if x in ['sí', 'si'] else 0)

# Se Agrupa por municipio
agrupado = df.groupby(['nombredepartamento', 'nombremunicipio']).agg(
    total=('matricula_Contratada', 'count'),
    con_matricula=('con_matricula_contratada', 'sum')
).reset_index()
agrupado['porcentaje_matricula'] = agrupado['con_matricula'] / agrupado['total'] * 100

#Se carga archivo GeoJSON local
geojson_path = 'https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/3aadedf47badbdac823b00dbe259f6bc6d9e1899/colombia.geo.json'
print(f"Attempting to read GeoJSON from: {geojson_path}") # Print the path being used
try:
    gdf = gpd.read_file(geojson_path)

    print("Columns in gdf:", gdf.columns)
except FileNotFoundError:
    print(f"Error: The file '{geojson_path}' was not found. Please provide the correct path to your GeoJSON file.")

# SE inspecciona las columnas y se remplaza dpto y mpio con el nombre corrrecto
gdf['DPTO'] = gdf['DPTO'].str.lower().str.strip()
gdf['NOMBRE_DPT'] = gdf['NOMBRE_DPT'].str.lower().str.strip()


# Se Unen con los datos
gdf_merged = gdf.merge(agrupado, left_on=['DPTO', 'NOMBRE_DPT'], right_on=['nombredepartamento', 'nombremunicipio'], how='left')
gdf_merged['porcentaje_matricula'] = gdf_merged['porcentaje_matricula'].fillna(0)

# Se visualiza
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
gdf_merged.plot(column='porcentaje_matricula',
                cmap='coolwarm',
                linewidth=0.3,
                edgecolor='grey',
                legend=True,
                legend_kwds={'label': "Porcentaje con Matrícula Contratada"},
                ax=ax)
plt.title("Presencia de Matrícula Contratada por Municipio y Departamento", fontsize=16)
plt.axis("off")
plt.show()

# Se carga el GeoJSON file for departmentos
url_geojson = 'https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/3aadedf47badbdac823b00dbe259f6bc6d9e1899/colombia.geo.json'
gdf_departamentos = gpd.read_file(url_geojson)

# Se aseguran las coincidencia en nombres de departamentos
gdf_departamentos['NOMBRE_DPT'] = gdf_departamentos['NOMBRE_DPT'].str.upper()
df['nombredepartamento'] = df['nombredepartamento'].str.upper()

# Se normalizan la columna de matrícula contratada
df['matricula_Contratada'] = df['matricula_Contratada'].fillna('No').str.strip().str.lower()
df['presencia_contratada'] = df['matricula_Contratada'].apply(lambda x: 1 if x in ['sí', 'si'] else 0)

# Se Agrupa por departamento para saber si hay al menos una institución con matrícula contratada
resumen = df.groupby('nombredepartamento')['presencia_contratada'].max().reset_index()

# Se Une con el GeoDataFrame
gdf_mapa = gdf_departamentos.merge(resumen, how='left', left_on='NOMBRE_DPT', right_on='nombredepartamento')

# Se Reemplaza NaN por 0
gdf_mapa['presencia_contratada'] = gdf_mapa['presencia_contratada'].fillna(0)

# Se crea mapa base
m = folium.Map(location=[4.5709, -74.2973], zoom_start=5, tiles='CartoDB positron')

# Se Añade capa de coropletas
folium.Choropleth(
    geo_data=gdf_mapa,
    name='choropleth',
    data=gdf_mapa,
    columns=['NOMBRE_DPT', 'presencia_contratada'],
    key_on='feature.properties.NOMBRE_DPT',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name='Presencia de Matrícula Contratada (1: sí, 0: no)',
    nan_fill_color='lightgray',
).add_to(m)

# Se Añade etiquetas emergentes
for _, row in gdf_mapa.iterrows():
    popup_text = f"{row['NOMBRE_DPT']}<br>¿Matrícula contratada?: {'Sí' if row['presencia_contratada'] == 1 else 'No'}"
    folium.Popup(popup_text).add_to(
        folium.GeoJson(row['geometry'], style_function=lambda x: {'fillOpacity': 0})
    )

# Se visualiza mapa
m
```

Objetivo específico:
Agrupar los establecimientos educativos del país a partir de la similitud en variables territoriales e institucionales —como zona, tipo de establecimiento, tipo de prestador y departamento— con el fin de analizar la composición interna de cada clúster, identificar posibles patrones de agrupamiento regional o administrativo, y examinar si ciertos clústeres presentan mayor asociación con la presencia de matrícula contratada.

```python
# Se cargan las librerias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python

# Cargar los datos limpios desde el archivo CSV
df_model = pd.read_csv("/cleaned_ESTABLECIMIENTOS_EDUCATIVOS-COLOMBIA_20250605 (1) (6).csv")
df_model.columns = df_model.columns.str.strip()

#se muestra columnas y filas para confirmar que la data cargo correctamente
display(df_model.head())
display(df_model.columns)
```

```python
# Se Limpian los nombres de columnas
df_model.columns = df_model.columns.str.strip()

#Se establecen Variables categóricas a usar
X = df_model[['zona', 'tipo_Establecimiento', 'prestador_de_Servicio', 'nombredepartamento']]
X_encoded = X.apply(LabelEncoder().fit_transform)

# Se Aplica el modelo KMeans , se eligieron  2 clústeres
kmeans = KMeans(n_clusters=2, random_state=42)
df_model['cluster'] = kmeans.fit_predict(X_encoded)

# Se utiliza PCA para Visualizar con el mismo y se reduce a 2 dimensiones
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_encoded)

# se realiza la visualizacion de los clustering

from matplotlib.colors import ListedColormap # Import ListedColormap
colores = ListedColormap(['#003f5c', '#ff69b4'])  # Azul oscuro y rosado

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df_model['cluster'], cmap=colores, alpha=0.6)
plt.title('Clustering de establecimientos educativos (K-Means)')
plt.xlabel('Componente PCA 1')
plt.ylabel('Componente PCA 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
# Contar cantidad de registros por clúster y tipo de matrícula
cruce = df_model.groupby(['cluster', 'matricula_binaria']).size().unstack(fill_value=0)

# Calcular proporciones por clúster
proporciones = cruce.div(cruce.sum(axis=1), axis=0)

print("Distribución de matrícula por clúster:\n")
print(proporciones)

# Se crea grafico
import matplotlib.pyplot as plt
colores = ['#003f5c', '#ff69b4']
proporciones.plot(kind='bar', stacked=True, color=colores)
plt.title("Proporción de matrícula contratada por clúster")
plt.xlabel("Clúster")
plt.ylabel("Proporción")
plt.legend(title="Matrícula contratada", labels=["No", "Sí"])
plt.tight_layout()
plt.grid(False)
plt.show()
```
